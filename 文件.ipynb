指针位置

```python
# 1. 打开文件
file = open("README")

# 2. 读取文件内容
text = file.read()
print(text)
print(len(text))

print("-" * 50)

text = file.read()
print(text)
print(len(text))

print("-" * 50)

#指针已经读到最后，调用seek方法设置指针位置 .seek（offset，whence=0）偏移量offset，whence0头部1当前2尾部
file.seek(0,0)
text = file.read()
print(text)
print(len(text))

# 3. 关闭文件
file.close()

```

读写

```python
file_read=open("README")
file_write=open("file.txt",'w')
while(1):
    text=file_read.readline()
    if not text:
        break;
    file_write.write(text)

file_read.close()
file_write.close()

#.write(str)
#.writeline(list)
#.read()
#.readline()
```
