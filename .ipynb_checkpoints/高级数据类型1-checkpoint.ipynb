list

```python
name_list = ["zhangsan", "lisi", "wangwu"]
```

```python
print(name_list.index("wangwu"))#得到序号
```

```python
name_list.append("王小二")#列表末尾添加元素
print(name_list)
```

```python
name_list.insert(1, "小美眉")#指定位置插入元素
print(name_list)
```

```python
temp_list = ["孙悟空", "猪二哥", "沙师弟"]
name_list.extend(temp_list)#拼接列表
print(name_list)
```

```python
name_list.remove("wangwu")#删除元素
print(name_list)
```

```python
name_list.pop()#删除最后一个元素
print(name_list)
```

```python
name_list.pop(3)#删除指定位置元素
print(name_list)
```

```python
list_len = len(name_list)#长度
print("列表中包含 %d 个元素" % list_len)
```

```python
count = name_list.count("张三")#计数
print("张三出现了 %d 次" % count)
```

```python
num_list = [6, 8, 4, 1, 10]
num_list.reverse()#列表反转
print(num_list)
```

```python
for each in num_list:
    print(each)
```

tuple

```python
info_tuple = ("zhangsan", 18, 1.75, "zhangsan")

# 1. 取值和取索引
print(info_tuple[0])
# 已经知道数据的内容，希望知道该数据在元组中的索引
print(info_tuple.index("zhangsan"))

# 2. 统计计数
print(info_tuple.count("zhangsan"))
# 统计元组中包含元素的个数
print(len(info_tuple))
```

```python
for each in info_tuple:
    print(each)
```

```python
info_tuple = ("小明", 21, 1.85)
# 格式化字符串后面的 `()` 本质上就是元组
info_str = "%s 年龄是 %d 身高是 %.2f" % info_tuple
print(info_str)
```

dictionary

```python
xiaoming_dict = {"name": "小明"}
xiaoming_dict.pop("name")#删除键值对
print(xiaoming_dict)
```

```python
xiaoming_dict = {"name": "小明",
                 "age": 18}
```

```python
# 1. 统计键值对数量
print(len(xiaoming_dict))
```

```python
# 2. 合并字典
temp_dict = {"height": 1.75,
             "age": 20}

# 注意：如果被合并的字典中包含已经存在的键值对，会覆盖原有的键值对
xiaoming_dict.update(temp_dict)
print(xiaoming_dict)
```

```python
# 3. 清空字典
xiaoming_dict.clear()
print(xiaoming_dict)
```

```python
xiaoming_dict = {"name": "小明",
                 "qq": "123456",
                 "phone": "10086"}
keys=list(xiaoming_dict.keys())
print(keys)
```

```python
# 迭代遍历字典
# 变量k是每一次循环中，获取到的键值对的key
for k in xiaoming_dict:
    print("%s-%s"%(k,xiaoming_dict[k]))
```

```python
card_list = [
    {"name": "张三",
     "qq": "12345",
     "phone": "110"},
    {"name": "李四",
     "qq": "54321",
     "phone": "10086"}
]

for card_info in card_list:

    print(card_info)
```

str

```python
str1 = "hello python"
for char in str1:
    print(char)
```

```python
hello_str = "hello hello"

# 1. 统计字符串长度
print(len(hello_str))

# 2. 统计某一个小（子）字符串出现的次数
print(hello_str.count("llo"))
print(hello_str.count("abc"))

# 3. 某一个子字符串第一次出现的位置
print(hello_str.index("llo"))

# 注意：如果使用index方法传递的子字符串不存在，程序会报错！
# print(hello_str.index("abc"))
```

```python
# 1. 判断空白字符
space_str = "      \t\n\r"

print(space_str.isspace())

# 2. 判断字符串中是否只包含数字
# 1> 都不能判断小数
# num_str = "1.1"
# 2> unicode 字符串
# num_str = "\u00b2"
# 3> 中文数字
num_str = "一千零一"

print(num_str)
print(num_str.isdecimal())
print(num_str.isdigit())
print(num_str.isnumeric())
```
