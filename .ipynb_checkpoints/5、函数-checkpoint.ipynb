# 嵌套调用

```python
def print_line(char, times):

    """
    打印单行分隔线
    :param char: 分隔字符
    :param times: 重复次数
    """
    print(char * times)


def print_lines(char, times ,rows):

    """
    打印多行分隔线
    :param char: 分隔线使用的分隔字符
    :param times: 分隔线重复的次数
    """
    row=0;
    while row < rows:
        print_line(char, times)
        row += 1

print_lines("-", 20,5)
```

# 收集参数

```python
#Python 中函数参数带*的话，将会收集非关键字的参数到一个元组中；
def print_params(*params):
    print(params)
print_params("hello")
print_params(1,2,3)
```

```python
def print_params(title,*params):
    print(title)
    print(params)
print_params("hello",1,2,3)
print_params("hello")
```

```python
#带*参数放在中间位置时，后面的参数在传入时必须指定参数名
def fun(x,*y,z):
    print(x,y,z)
fun(1,2,3,4,5,z=6)
```

```python
#Python 中函数参数带**的话，将会手机关键字参数到一个字典中；
def fun(**params):
    print(params)
fun(x=1,y=2,z=3)
```

```python
def fun(x,y,z=3,*pospar,**keypar):
    print(x,y,z)
    print(pospar)
    print(keypar)
fun(1,2,3,4,5,6,a=11,b=22,c=33)
#z是带默认值的参数
```

```python
fun(1,1,1,4,5,6,a=11,b=22,c=33)
```

```python
fun(0,0)
```
